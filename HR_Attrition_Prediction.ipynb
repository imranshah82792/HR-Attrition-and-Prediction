```python
# Question:1 What is the Overall Attrition rate of the Company

# To see counts for YES/NO
# data['Attrition'].value_counts()

# Overall Attrition rate
total_emp = len(data)
left_emp = (data['Attrition'] == "Yes").sum()

att_rate =(left_emp / total_emp) * 100 
print(f"Overall Attrition rate{att_rate:.2f}%")
```

```python
# Simple visualization with matplotlib

att_counts = data['Attrition'].value_counts()
att_counts.plot(kind = "bar",color = ["steelblue","salmon"])
plt.title("Employee Attrition count")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)
plt.show()
```

```python
# Question:2 How is Attrition distibuted across departments

dep_att_count = data.groupby(['Department','Attrition']).size().unstack(fill_value = 0)

dep_att_count["total"] = dep_att_count.sum(axis = 1)
dep_att_count["att_rate%"] = dep_att_count["Yes"]/dep_att_count["total"] * 100
print(dep_att_count)
```

```python

plt.figure(figsize=(6,4))
plt.bar(dep_att_count.index, dep_att_count["att_rate%"], color="teal")
plt.ylabel("Attrition rate (%)")
plt.title("Attrition rate by Department")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
# Question:3 Does Attrition differ by gender

gen_att = pd.crosstab(data['Gender'] , data['Attrition'])

gen_att['Total'] = gen_att.sum(axis = 1)
gen_att['att_rate%'] = gen_att['Yes'] / gen_att['Total'] * 100
print(gen_att)
```

```python
# Question:4 What is the age distribution of employees who leaves vs stay

age_by_att = data.groupby('Attrition')['Age'].describe()
print(age_by_att)
```

```python
import matplotlib.pyplot as plt

left  = data[data["Attrition"] == "Yes"]["Age"]
stay  = data[data["Attrition"] == "No"]["Age"]

plt.hist(stay, bins=10, alpha=0.6, label="Stay", color="skyblue")
plt.hist(left, bins=10, alpha=0.6, label="Leave", color="salmon")

plt.xlabel("Age")
plt.ylabel("Number of employees")
plt.title("Age distribution: Leave vs Stay")
plt.legend()
plt.show()
```

```python
# Question:5 How does monthly income relate to attrition across job roles

inc_role_att = (data.groupby(['JobRole','Attrition'])['MonthlyIncome'].mean().unstack())
print(inc_role_att)
```

```python
# Question:6 Is there a relationship b/w distance from home and Attrition

dist_stat = data.groupby('Attrition')['DistanceFromHome'].describe()
print(dist_stat)
```

```python

data.boxplot(column="DistanceFromHome", by="Attrition")
plt.title("Distance From Home vs Attrition")
plt.suptitle("")  # remove automatic extra title
plt.ylabel("Distance From Home")
plt.show()
```

```python
# Question:7 How do overtime and work_life balance influence attrition


pivot = pd.crosstab(
    index=[data["OverTime"], data["WorkLifeBalance"]],
    columns=data["Attrition"]
)

pivot["Total"] = pivot.sum(axis=1)
pivot["AttritionRate_%"] = pivot["Yes"] / pivot["Total"] * 100

print(pivot)
```

```python
# Question:8 Do job satisfaction and reletionship satisfaction impact attrition

# Job satisfaction vs attrition
# js = pd.crosstab(data["JobSatisfaction"], data["Attrition"])
# js["Total"] = js.sum(axis=1)
# js["AttritionRate_%"] = js["Yes"] / js["Total"] * 100
# print(js)


# Relationship satisfaction vs attrition

rs = pd.crosstab(data["RelationshipSatisfaction"], data["Attrition"])
rs["Total"] = rs.sum(axis=1)
rs["AttritionRate_%"] = rs["Yes"] / rs["Total"] * 100
print(rs)
```

```python
# Which top 5 features in this dataset are most associated with attrition?


data["Attrition_Flag"] = (data["Attrition"] == "Yes").astype(int)
df_num = pd.get_dummies(data.drop(columns=["Attrition"]), drop_first=True)


corr = df_num.corr()["Attrition_Flag"].drop("Attrition_Flag")

top5 = corr.reindex(corr.abs().sort_values(ascending=False).index)[:5]
print(top5)

```
